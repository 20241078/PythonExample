```python
## 1. f-string, format 예제
price = 59.3524
txt = f"The price is {price:.2f} dollars"
print(txt)
```

```python
## 2. 정수나 string도 T, F 판단 가능. 0이면 F
x = "Hello"
y = 15
print(bool(x))
print(bool(y))
```

```python
## 3. 
numbers = [1, 2, 3, 4, 5]
count = len(numbers)
if count > 3:
    print(f"List has {count} elements")

if (count := len(numbers)) > 3:
    print(f"List has {count} elements")
```

```python
## 4. Python은 < 한 번에 쓸 수 있음
x = 5
print(1 < x < 10)
print(1 < x and x < 10)
```

```python
## 5. 계산 순서
print((6 + 3) - (6 + 3))
print(100 + 5 * 3)
print(5 + 4 - 7 + 3)
```

```python
### 6. 이진수 비교해서 정수로 나옴
print(6 & 3)
print(6 | 3)
print(6 ^ 3)
```

```python
## 7. hello는 소문자니까 F
text = "Hello World"
print("H" in text)
print("hello" in text)
print("z" not in text)
```

```python
## 8. in: fruits안에 바나나가 있는지
fruits = ["apple", "banana", "cherry"]
print("banana" in fruits)
```

```python
## 9. is : 값, 메모리, 타입을 비교 // == : 값만 비교
x = ["apple", "banana"]
y = ["apple", "banana"]
z = x
print(x is z)
print(x is y)
print(x == y)
```

```python
## 10. 기호. floor devision을 Python에서 제공함
x = 15
y = 4
print(x + y)
print(x - y)
print(x * y)
print(x / y)
print(x % y)
print(x ** y)
print(x // y)
```
